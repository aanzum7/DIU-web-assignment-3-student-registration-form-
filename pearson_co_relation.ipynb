```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
CustomerRelation = pd.read_csv("/PearsonV3.csv")
```

```python
CustomerRelation.head()
```

```python
pearsoncorr = CustomerRelation.corr(method='pearson')
pearsoncorr
```

```python
import seaborn as sb
```

```python
import matplotlib.pyplot as pl
```

```python
plt.figure(figsize=(12, 9))
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```
